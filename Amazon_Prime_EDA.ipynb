## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Writing code to make your dataset analysis ready.

#Removing duplicate rows from both datasets
data1.drop_duplicates(inplace = True)
data2.drop_duplicates(inplace = True)
```

```python
#merge 2 datasets on common column id
data = pd.merge(data1,data2, on = 'id')
data.head()
```

```python
#check for null values
data.isnull().sum()
```

```python
# Dropping the 'description' column because it does not contribute meaningful categorical or numerical features for EDA,
# and no useful insights can be quantitatively extracted from it.
data=data.drop(columns=['description'])
```

```python
# Filling missing values in 'age_certification' with 'Unknown' since over 50% of the entries are null.
# Using 'Unknown' avoids introducing bias that would occur if we used the mode,
# and helps retain the completeness of the column for EDA.
data['age_certification'].fillna('Unknown', inplace=True)
```

```python
#116026 records are of type movie so seasons column will have 116026 null entries since movies don't have seasons
data['type'].value_counts()
```

```python
#Movies don't have seasons so these entries would naturally be null.
#Instead of removing these rows, replace them with 0.
data['seasons'].fillna(0, inplace=True)
```

```python
# Dropping 'imdb_id' since it is mostly a reference field and contains many nulls,
# and does not contribute meaningful information to the EDA.
data.drop(columns=['imdb_id'], inplace=True)
```

```python
#'imdb_score' represents the rating of a Movie or TV show
#Removing null values in 'imdb_score' might lead to data loss it's better to replace them with mean.
data['imdb_score'].fillna(round(data['imdb_score'].mean(),1), inplace=True)
```

```python
#'imdb_votes' represents the number of votes a Movie or TV show has recieved.
#Missing values in 'imdb_votes' may indicate movie is relatively new or has not gained much attention.
#It's better to replace the null values with 0.
data['imdb_votes'].fillna(0, inplace = True)
```

```python
# replace null values in 'tmdb_popularity' column with mean
data['tmdb_popularity'].fillna(round(data['tmdb_popularity'].mean(),2), inplace =True)
```

```python
# Imputing missing values in 'tmdb_score' using KNNImputer.
# The column has a large number of missing entries, and simple techniques
# like mean/median imputation would distort the rating distribution.
# KNNImputer estimates missing scores based on similarity with other titles
# (e.g., using tmdb_popularity and related features), resulting in more
# realistic and data-driven values. This helps preserve patterns in the
# dataset without introducing bias.

from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)
data['tmdb_score'] = imputer.fit_transform(data[['tmdb_score', 'tmdb_popularity']])
```

```python
#Missing values in the 'characters' column indicate that the information about the character played by a person in movie or TV show is not available in the dataset. So replace them by 'unknown'

data['character'].fillna('unknown', inplace = True)
```
